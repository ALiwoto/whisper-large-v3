```python
import torch
from transformers import pipeline

pipe = pipeline("automatic-speech-recognition",
               "openai/whisper-large-v3",
               torch_dtype=torch.float16,
               device="cuda:0")
```

```python
pipe_type = type(pipe)

print(f"pipe type is {pipe_type}")
```

```python
pipe("https://cdn-media.huggingface.co/speech_samples/sample1.flac")
```

```python
# if we want to convert from other formats (e.g. m4a) to mp3, we can just use this command:
!ffmpeg -i sample3.m4a -c:v copy -c:a libmp3lame -q:a 4 sample3.mp3
```

```python

sample_result = pipe("./sample.mp3")
print(sample_result)

# create a file called output.txt and write sample_result['text'] inside of it
with open('output.txt', 'w') as f:
    f.write(sample_result['text'])

# later on, you can just download the file from file panel in Colab
```
